# Artificial Bee Colony

```python
def array_(aba):
    aba_=np.array(aba)
    return aba_
```

```python
g_best_aba=[]
total_time=[]
for i in range(len(func)):
    start = time.time()  
    aba = SwarmPackagePy.aba(40, func[i], minmaxlist[i][0],minmaxlist[i][1],2,1000)
    aba_best=aba.get_Gbest()
    aba_points = aba.get_agents()
    aba_points = [[func[i](array_(j)) for j in agent] for agent in aba_points]
    aba_best_=array_(aba_best)
    end = time.time()  
    running_time = end - start                   
    print('\n')
    print(f'function_name = {func_name[i]}') 
    print(f'global_best_position = {aba_best}') 
    print(f'global_best = {func[i](aba_best_)}') 
    print(f'Running Time : {int(np.round(running_time,0))} second')
    g_best_aba.append(np.min(aba_points, axis=1))
    total_time.append(running_time)
print('\n')
print(f'total time of Artificial Bee Colony = {int(np.round(np.sum(total_time),0))} Second') 
```

# Fireflyalgorithm

```python
import numpy as np
from numpy.random import default_rng

class FireflyAlgorithm:
    def __init__(self, n_fireflies=40, alpha=1.0, betamin=1.0, gamma=0.01, seed=None):
        self.n_fireflies = n_fireflies
        self.alpha = alpha
        self.betamin = betamin
        self.gamma = gamma
        self.rng = default_rng(seed)

    def Run_algorithm(self, function, dim, low, high, MaxGeneration):
        fireflies = self.rng.uniform(low, high, (self.n_fireflies, dim))
        intensity = np.apply_along_axis(function, 1, fireflies)
        best = np.min(intensity)
        t = self.n_fireflies
        new_alpha = self.alpha
        search_range = high - low
        list_g_best =[]
        
        while t <= MaxGeneration:
            new_alpha *= 0.97
            for i in range(self.n_fireflies):
                for j in range(self.n_fireflies):
                    if intensity[i] >= intensity[j]:
                        r = np.sum(np.square(fireflies[i] - fireflies[j]), axis=-1)
                        beta = self.betamin * np.exp(-self.gamma * r)
                        steps = new_alpha * (self.rng.random(dim) - 0.5) * search_range
                        fireflies[i] += beta * (fireflies[j] - fireflies[i]) + steps
                        fireflies[i] = np.clip(fireflies[i], low, high)
                        intensity[i] = function(fireflies[i])
                        t += 1
                        best = min(intensity[i], best)
                        list_g_best.append(best)
        return best, list_g_best
```

```python
FA = FireflyAlgorithm()
g_best_Fa=[]
total_time_fa=[]
for i in range(len(func)):
     start = time.time()  
     g_best_Firefly = FA.Run_algorithm(function=func[i], dim=2, low=minmaxlist[i][0], high= minmaxlist[i][1], MaxGeneration=1000)   
     end = time.time()  
     running_time = end - start                   
     print('\n')
     print(f'function_name = {func_name[i]}') 
     print(f'global_best = {g_best_Firefly[0]}') 
     print(f'Running Time : {(np.round(running_time,2))} second')
     g_best_Fa.append(g_best_Firefly[1])
     total_time_fa.append(running_time)
print('\n')
print(f'Total time of FireflyAlgorithm = {((np.sum(total_time_fa)))} Second') 
```

```python
plt.figure(figsize=(30,100))
for i in range(19):
    ax = plt.subplot(16, 3, i + 1)
    plt.plot(g_best_Fa[i][0:1000],color='b',label='FireflyAlgorithm')
    plt.xlabel("Number of iteration",fontsize=12);
    plt.ylabel("Fitness value",fontsize=12);
    plt.legend()
    plt.title("func_name={}".format(
        func_name[i],fontsize=15));
plt.show()
```

# Particle Swarm Optimization

```python
g_best_pso=[]
total_time_pso=[]
for i in range(len(func)):
    start = time.time()  
    pso = SwarmPackagePy.pso(40, func[i], minmaxlist[i][0],minmaxlist[i][1],2,1000, w=0.75, c1=1.7, c2=1.7) 
    pso_best=pso.get_Gbest()
    pso_points = pso.get_agents()
    pso_points = [[func[i](array_(j)) for j in agent] for agent in pso_points]
    pso_best_=array_(pso_best)
    end = time.time()  
    running_time = end - start                   
    print('\n')
    print(f'function_name = {func_name[i]}') 
    print(f'global_best_position = {pso_best}') 
    print(f'global_best = {func[i](pso_best_)}') 
    print(f'Running Time : {(np.round(running_time,2))} second')
    g_best_pso.append(np.min(pso_points, axis=1))
    total_time_pso.append(running_time)
print('\n')
print(f'Total time of Particle Swarm Optimization = {(np.sum(total_time_pso))} Second') 
```
